```python
# C√©lula 1 - Importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configura√ß√µes
%matplotlib inline
plt.style.use('ggplot')
sns.set_palette("Set2")
print("‚úÖ Bibliotecas importadas!")

# C√©lula 2 - Verificar se arquivo existe
print("üìÇ Verificando estrutura de pastas...")
print("Pasta atual:", os.getcwd())
print("Arquivos na pasta:", os.listdir('.'))

if os.path.exists('_data'):
    print("Conte√∫do da pasta _data:", os.listdir('_data'))
else:
    print("‚ùå Pasta _data n√£o encontrada!")
    print("‚ùå Crie a pasta _data e coloque o arquivo CSV dentro!")

# C√©lula 3 - Carregar dados 
try:
    # Tente primeiro com caminho relativo
    df = pd.read_csv('./_data/dataset_2021-5-26-10-14.csv', sep='\t', encoding='utf-8')
    print("‚úÖ Arquivo carregado com caminho relativo!")
except FileNotFoundError:
    try:
        # Se n√£o funcionar, use caminho ABSOLUTO (ALTERE PARA SEU CAMINHO)
        caminho_absoluto = r'C:\Users\cesar\OneDrive\√Årea de Trabalho\projeto-xhealth\_data\dataset_2021-5-26-10-14.csv'
        df = pd.read_csv(caminho_absoluto, sep='\t', encoding='utf-8')
        print("‚úÖ Arquivo carregado com caminho absoluto!")
    except Exception as e:
        print("‚ùå Erro ao carregar arquivo:", e)
        print("‚ùå Verifique o caminho do arquivo!")

# C√©lula 4 - Primeira visualiza√ß√£o
print("üìä Shape do dataset:", df.shape)
print("\nüîç Primeiras 5 linhas:")
display(df.head())

# C√©lula 5 - Informa√ß√µes b√°sicas
print("\nüìã Informa√ß√µes do dataset:")
df.info()

# C√©lula 6 - Tratar valores missing
df = df.replace('missing', np.nan)
print("\nüîç Valores nulos por coluna:")
print(df.isnull().sum())

# C√©lula 7 - Estat√≠sticas descritivas
print("\nüìà Estat√≠sticas descritivas:")
display(df.describe())

# C√©lula 8 - An√°lise da vari√°vel target
print("\nüéØ Distribui√ß√£o de default:")
print(df['default'].value_counts())

plt.figure(figsize=(8, 6))
df['default'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Distribui√ß√£o de Default (0 = Pago, 1 = Calote)')
plt.xlabel('Status do Pagamento')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.show()

```
